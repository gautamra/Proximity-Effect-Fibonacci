```python
from proximitylib import *

font = {'family' : 'sans-serif',
        'weight' : 'normal',
        'size'   : 14}

mpl.rc('font', **font)
mpl.rcParams['figure.figsize'] = 6,3.6
mpl.rcParams['figure.facecolor'] = "w"
```

```python
chain = simple_ring(chain1D(199, u = 0.75, v = -1.5, wu = 0), beta = 20)

DOS, eex = chain.get_DOS(gam = 0.06)
plt.plot(eex, chain.DOS, label = "0 T")
minE = np.abs(chain.evals).min()
plt.plot([-minE, minE],[0.2,0.2], 'C0')
plt.plot([-minE,-minE],[0.19,0.21], 'C0')
plt.plot([minE,minE],[0.19,0.21], 'C0')
# plt.annotate("$E_g$", (0,0.2),(-0.1, 0.24), arrowprops=dict(facecolor='black', shrink=0.05))
plt.xlabel("$\epsilon/t$")
plt.ylabel("$\\rho(\epsilon)$")



chain2 = simple_ring(chain1D(199, u = 0.75, v = -1.5, wu = 0), beta = 15)
DOS2, eex2 = chain2.get_DOS(gam = 0.06)
minE2 = np.abs(chain2.evals).min()
plt.plot([-minE2, minE2],[0.2,0.2], 'C1')
plt.plot([-minE2,-minE2],[0.19,0.21], 'C1')
plt.plot([minE2,minE2],[0.19,0.21], 'C1')

plt.plot(eex2, chain2.DOS, label = "Finite T ($\\beta$ = 15)")

plt.title("Length = {}".format(chain.LL))
plt.legend()
plt.show()
```

```python
# betas = [6,7,8, 10,12,12.5,13,14,15,16,17 ,20,25,30, 50, 100]
Ts = 1/100*np.arange(0.5,15,0.5)
betas = 1/Ts
Deltas = []
for beta in tqdm(betas):
    chain = simple_ring(chain1D(199, u = 0.75, v = -1.5, wu = 0), beta = beta)
    Deltas.append(chain.Delta[0])

plt.plot(1/np.array(betas), Deltas, ".")
plt.xlabel("$1/\\beta$")
plt.ylabel("$\Delta$")
```

```python
plt.plot(1/np.array(betas), Deltas, "o")
plt.xlabel("$1/\\beta$")
plt.ylabel("$\Delta$")
plt.annotate("$1/\\beta_c \\approx {}$".format(1/12.5),(1/betas[15],Deltas[15]), (1/betas[15],Deltas[15]+0.02), arrowprops=dict(arrowstyle="->"))
```
